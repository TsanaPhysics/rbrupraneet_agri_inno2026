**เป้าหมาย:** 
1. วิเคราะห์กราฟ Loss และ Accuracy เพื่อดูการเรียนรู้ของโมเดล
2. ทำนายผลภาพใหม่ๆ (Inference) ที่ AI ไม่เคยเห็นมาก่อน
3. ส่งออกโมเดลเป็นไฟล์ `.h5` หรือ `TFLite` เพื่อนำไปพัฒนาแอปพลิเคชัน

```python
# สมมติการพยากรณ์ผลลัพธ์ (Mock Inference)
import numpy as np

# รายชื่อหมวดหมู่ที่เราสอนไป
class_names = ['Normal (ปกติ)', 'Leaf Spot (ราใบติด)', 'Durian Red Rust (ราสนิมแดง)']

# ทายผลจากภาพใบไม้ใหม่ 1 ใบ
def predict_disease(prediction_array):
    idx = np.argmax(prediction_array)
    return class_names[idx], prediction_array[idx]

# จำลองผลลัพธ์จากโมเดล
mock_prediction = [0.05, 0.92, 0.03]
label, confidence = predict_disease(mock_prediction)

print(f"🔍 AI ผลสรุป: {label}")
print(f"🎯 ความมั่นใจ: {confidence * 100:.2f}%")
```
