# ด่านแรก: พิกเซลและสีสัน (Pixels & Colors)
## เปิดภาพด้วย OpenCV

```python
import cv2
import matplotlib.pyplot as plt

# โหลดภาพใบไม้และสลับค่าสี BGR สู่ RGB ให้สีสมจริง
img = cv2.imread('leaf.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Original Leaf')
plt.show()
```
