ในตอนนี้เราจะมาประกอบร่าง "โครงข่ายประสาทเทียม" (CNN) ซึ่งเป็นหัวใจสำคัญที่ทำให้คอมพิวเตอร์สามารถจำแนกโรคทุเรียนได้เหมือนมีผู้เชี่ยวชาญมานั่งดูเองเลยครับ

**เป้าหมาย:** 
1. เข้าใจเลเยอร์หลักของ CNN (Convolution, Pooling, Dense)
2. ก่อสร้างโมเดลด้วย `TensorFlow / Keras` แบบ Step-by-Step
3. เรียนรู้วิธีการกำหนด Output ให้ตรงกับจำนวนโรคที่เราต้องการตรวจจับ

```python
from tensorflow.keras import layers, models

# สร้างสถาปัตยกรรมทางปัญญา (Building the Brain)
model = models.Sequential([
    # ชั้นที่ 1: การดึงฟีเจอร์เด่น (ขอบใบ, จุดโรค)
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    
    # ชั้นที่ 2: การเจาะลึกรายละเอียด
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    # ชั้นที่ 3: การสรุปและจัดหมวดหมู่
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='softmax') # เราจะตรวจ 3 อย่าง: ปกติ, ราใบติด, ทุเรียนแดง
])

print("🏗️ โครงสร้างสมองกล (CNN) สร้างเสร็จเรียบร้อย!")
model.summary()
```

## 1. การเตรียมการฝึกฝน (Compilation)
เราต้องกำหนด "กฎเกณฑ์การเรียนรู้" ให้กับสมองของเรา ว่าจะวัดผลสำเร็จอย่างไร และจะปรับปรุงตัวเองอย่างไรครับ

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("📖 พร้อมเริ่มต้นการเรียนรู้ (Ready to Train)")
```
