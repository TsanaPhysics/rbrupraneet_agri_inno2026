# AI วิเคราะห์คุณภาพดิน (NPK & pH)
## อัลกอริทึมจัดการค่า pH

```python
# แปลงค่าแรงดันไฟฟ้าจากหัวโพรบเครื่องแก้วเป็นค่า pH 0-14
def calculate_ph(voltage):
    # สมการ Calibration: pH = 7 + ((V_neutral - V_read) / V_slope)
    return 7 + ((2.5 - voltage) / 0.18)

ph_val = calculate_ph(2.8)
print(f'🧪 ค่ากรด-ด่างที่ตรวจวัดได้: {ph_val:.2f} pH')
```
