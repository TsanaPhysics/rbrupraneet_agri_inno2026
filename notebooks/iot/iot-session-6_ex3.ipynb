# โปรโตคอลเพื่อการเกษตร (MQTT 101)
## ตั้งตารอคำสั่ง (Subscribe)

```python
def on_message_received(topic, msg):
    print(f'📩 มีโทรจิตส่งมาจากคลาวด์หัวข้อ {topic}: คำสั่ง {msg}')
    
client.set_callback(on_message_received)
# สมัครรอรับข่าวสารเฉพาะรหัสต้นไม้ของเรา
client.subscribe(b'rbru/farm/cmd/node1')
print('🎧 กำลังดักฟังคำสั่งบนคลาวด์...')
```
