## แปลงรหัสหยาบสู่เปอร์เซ็นต์ (Map Function)

```python
def map_percent(val, in_min, in_max):
    # สเกล 0-4095 แปลงเป็น 0-100%
    return (val - in_min) * 100 / (in_max - in_min)

moisture = map_percent(raw_value, 0, 4095)
print(f'💧 ความชื้นสัมพัทธ์ในดิน: {moisture:.1f}%')
```
