## ระบบป้องกันน้ำแห้งปั๊มระเบิด (Hardware Failsafe)

```python
water_level_sensor = Pin(35, Pin.IN)

# ห้ามสั่งรดน้ำถ้าในแทงก์ไม่มีน้ำ (Sensor กลับค่าเป็นตัวเลข)
if water_level_sensor.value() == 1:
    print('⚠️ ปั๊มน้ำทำงานไม่ได้ น้ำในแทงก์หมด!')
else:
    water_plant(10)
```
