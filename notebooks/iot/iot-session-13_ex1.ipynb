# AI วิเคราะห์คุณภาพดิน (NPK & pH)
## อ่านค่า NPK ผ่าน Modbus (RS485)

```python
# เซ็นเซอร์ NPK เกรดอุตสาหกรรมมักใช้โปรโตคอล RS485
from machine import UART

uart = UART(2, baudrate=9600, tx=17, rx=16)
# ส่งคำสั่ง Query ไปยังเซ็นเซอร์ NPK
query = b'\x01\x03\x00\x1e\x00\x03\x65\xcd'
uart.write(query)

print('⏳ กำลังสกัดค่า N-P-K จากหัวโพรบใต้ดิน...')
```
