# โปรโตคอลเพื่อการเกษตร (MQTT 101)
## สร้างตัวตนลงทะเบียน Cloud

```python
from umqtt.simple import MQTTClient
import ubinascii
import machine

# ดึงรหัสประจำตัวฮาร์ดแวร์ที่ไม่ซ้ำกันในโลก
CLIENT_ID = ubinascii.hexlify(machine.unique_id())
BROKER_IP = 'broker.hivemq.com'

client = MQTTClient(CLIENT_ID, BROKER_IP)
client.connect()
print('🔗 ผูกวิญญาณเข้ากับศูนย์สั่งการคลาวด์สำเร็จแล้ว!')
```
