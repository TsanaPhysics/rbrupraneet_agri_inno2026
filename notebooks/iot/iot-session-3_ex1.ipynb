# เซ็นเซอร์สภาพแวดล้อม (I2C Sensors)
## แสกนหาอุปกรณ์บน I2C (Bus Scan)

```python
from machine import I2C

# เปิด Bus หมายเลข 0 ด้วยความเร็ว 400kHz
i2c = I2C(0, scl=Pin(22), sda=Pin(21), freq=400000)

# สแกนหาที่อยู่ (Address) ของเซ็นเซอร์ในสาย
devices = i2c.scan()
print('พบอุปกรณ์ที่ตำแหน่ง:', [hex(d) for d in devices])
```
