# สมองกลตัดสินใจเบื้องต้น (Edge Rules Engine)
## การเขียน Logic ฝังบนชิป (Local Thresholds)

```python
class EdgeEngine:
    def __init__(self):
        self.TEMP_MAX = 35.0
        self.DRY_CRITICAL = 30.0
        
    def evaluate(self, current_temp, current_humid):
        if current_temp > self.TEMP_MAX and current_humid < self.DRY_CRITICAL:
            return 'ฉุกเฉิน: ร้อนและแห้งจัด สั่งสเปรย์หมอกทันที'
        return 'สถานการณ์ปกติ'
```
