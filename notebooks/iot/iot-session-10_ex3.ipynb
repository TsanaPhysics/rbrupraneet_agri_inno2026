# สลีปโหมดเซฟโลก (Deep Sleep Mastery)
## วิธีการปลุกจากเซ็นเซอร์ (Ext0 Wakeup)

```python
import esp32

# ตั้งค่าให้ ESP32 หลับอยู่ แต่ถ้า 'ปุ่มฉุกเฉิน' (หรือเซ็นเซอร์ขโมย) ถูกกด ให้ตื่นทันที!
wake_pin = Pin(14, mode=Pin.IN, pull=Pin.PULL_UP)
esp32.wake_on_ext0(pin=wake_pin, level=esp32.WAKEUP_ALL_LOW)

print('เซ็ตระบบปลุกเมื่อมีคนบุกรุก... เข้าสู่การจำศีลแล้ว!')
```
