# รักษาความลับการค้า (IoT Edge Security)
## เข้ารหัสเซ็นเซอร์ (Hashing)

```python
import uhashlib
import ubinascii

data_str = 'pump:on,secret:1234'
# ปั่นพาสเวิร์ดให้กลายเป็นรหัสยาวๆ หมุนย้อนกลับไม่ได้
hash_obj = uhashlib.sha256(data_str.encode())
encrypted = ubinascii.hexlify(hash_obj.digest())

print(f'🔐 รหัสตู้เซฟที่คนอื่นอ่านไม่ออก: {encrypted}')
```
