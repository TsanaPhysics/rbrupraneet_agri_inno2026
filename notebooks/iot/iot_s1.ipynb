## 1. อะไรคือ MicroPython?
MicroPython คือภาษา Python เวอร์ชันจิ๋วที่ออกแบบมาเพื่อทำงานบนคอมพิวเตอร์ขนาดเล็ก (Microcontrollers) โดยตรง

```python
from machine import Pin
import time

# กำหนดขา (Pin) ของ LED (ปกติ ESP32 จะมี LED อยู่ที่ขา 2)
led = Pin(2, Pin.OUT)

print("🟢 เริ่มต้นระบบตรวจสอบเซ็นเซอร์...")

# สั่งให้ไฟกระพริบ 5 ครั้ง
for i in range(5):
    led.on()  # เปิดไฟ
    print("💡 LED ON")
    time.sleep(0.5)
    led.off() # ปิดไฟ
    print("🌑 LED OFF")
    time.sleep(0.5)

print("✅ จบการทดสอบอุปกรณ์เบื้องต้น")
```
