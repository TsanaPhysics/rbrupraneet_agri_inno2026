## 1. JSON คืออะไร?
JSON (JavaScript Object Notation) คือรูปแบบการส่งข้อมูลที่อ่านง่ายทั้งคนและคอมพิวเตอร์

```python
# ตัวอย่างข้อมูล JSON ของสถานีจันทบุรี
station_data = {
    "station_id": "48480",
    "location": "Chanthaburi",
    "sensors": [
        {"type": "temperature", "value": 30.5},
        {"type": "humidity", "value": 78.0}
    ]
}

print(f"พิกัดสถานี: {station_data['location']}")
print(f"อุณหภูมิที่วัดได้: {station_data['sensors'][0]['value']} °C")
```

## 2. การดึงข้อมูลด้วย `requests`
เราจะใช้เครื่องมือที่มีชื่อว่า `requests` ในการส่งคำขอไปยัง API Server เพื่อขอข้อมูลสภาพอากาศครับ

```python
import requests

# ตัวอย่าง URL ของ API (สมมติ)
API_URL = "https://data.tmd.go.th/api/v1/daily?station_id=48480"

print("📡 กำลังเชื่อมต่อกับ Server กรมอุตุนิยมวิทยา...")

# ในการใช้งานจริง เราต้องมี API Key
# response = requests.get(API_URL, headers={'Authorization': 'Bearer YOUR_KEY_HERE'})

# จำลองการดึงข้อมูลสกัดจาก API
mock_response = {
    "status": 200,
    "weather": "Scattered Rain",
    "rainfall_24h": 15.2,
    "wind_kmh": 12.5
}

if mock_response['status'] == 200:
    print("✅ เชื่อมต่อสำเร็จ!")
    print(f"สภาพอากาศ: {mock_response['weather']}")
    print(f"ปริมาณฝนสะสม: {mock_response['rainfall_24h']} มม.")
else:
    print("❌ การเชื่อมต่อขัดข้อง!")
```
