## 1. ตัวแปรและความหมาย (Variables in Agriculture)
เปรียบเสมือน "ถัง" ที่ใช้เก็บข้อมูลฟาร์มของเรา เช่น อุณหภูมิ, จำนวนต้นไม้, หรือชื่อพันธุ์ทุเรียน

```python
# ลองกำหนดข้อมูลสวนทุเรียนของเรา
farm_name = "RBRU Innovation Garden"
tree_count = 150 # จำนวนต้นทุเรียน (Integer)
current_temp = 32.5 # อุณหภูมิปัจจุบัน (Float)
is_raining = False # ฝนตกไหม? (Boolean)

print(f"หมู่บ้านเกษตร: {farm_name}")
print(f"เรามีต้นทุเรียนทั้งหมด: {tree_count} ต้น")
```

## 2. การคำนวณดัชนีความร้อน (Agri-Math Challenge)
ต้นทุเรียนอ่อนจะได้รับผลกระทบจากความชื้นและอุณหภูมิที่สูงเกินไป เรามาสร้างสูตรคำนวณ **Simple Heat Index** กันครับ

**สูตร:** `Heat Index = อุณหภูมิ + (0.05 * ความชื้น)`

```python
temp = 34.0    # องศาเซลเซียส
humidity = 85  # เปอร์เซ็นต์ความชื้น

# เขียนโปรแกรมคำนวณ
heat_index = temp + (0.05 * humidity)

print(f"อุณหภูมิที่วัดได้: {temp} °C")
print(f"ความรู้สึกร้อนจริง (Heat Index): {heat_index:.2f} °C")

# ระบบตัดสินใจ (Logic & Decision)
if heat_index > 36:
    print("❌ คำเตือน: อากาศร้อนจัด! โปรดเปิดระบบพ่นหมอกเพื่อลดอุณหภูมิในสวน")
else:
    print("✅ สภาพอากาศปกติ: ต้นทุเรียนยังอยู่ในสภาวะที่เหมาะสม")
```
