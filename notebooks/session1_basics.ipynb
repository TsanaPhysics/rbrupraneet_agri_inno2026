```python
# โปรแกรมคำนวณดัชนีความร้อน (Heat Index) แบบง่าย
temp = 32.5
humidity = 80

# ตรรกะการวิเคราะห์
heat_index = temp + (0.05 * humidity)

print(f"อุณหภูมิปัจจุบัน: {temp} °C")
print(f"ความรู้สึกจริง: {heat_index:.2f} °C")

if heat_index > 35:
    print("แจ้งเตือน: อากาศร้อนเกินไปสำหรับทุเรียนอ่อน!")
```
