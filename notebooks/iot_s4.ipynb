ในเซสชันสุดท้ายนี้ เราจะนำองค์ประกอบทั้งหมด (ฮาร์ดแวร์ + เซ็นเซอร์ + คลาวด์) มาเขียนอัลกอริทึมที่ทำให้ฟาร์มของเรา "รดน้ำเองได้" โดยใช้ตรรกะที่ชาญฉลาดและคำนวณจากสถานการณ์จริงครับ

**เป้าหมาย:** 
1. เขียนโปรแกรมควบคุมปั๊มน้ำ (Relay Control)
2. พัฒนาลอจิกการตัดสินใจแบบ Multi-Sensor (ความชื้นดิน + การพยากรณ์อากาศ)
3. สร้างระบบแจ้งเตือนผ่าน LINE Notify เมื่อเครื่องรดน้ำทำงาน

```python
from machine import Pin

# กำหนดขาควบคุมปั๊มน้ำ (Relay)
pump = Pin(27, Pin.OUT)

def auto_irrigation(moisture_val, rain_forecast_mm):
    """
    อัลกอริทึมการตัดสินใจรดน้ำอัจฉริยะ
    """
    print(f"🔍 ตรวจสอบสภาวะ: ดิน ({moisture_val}%) | ฝนพยากรณ์ ({rain_forecast_mm} mm)")
    
    if moisture_val < 40:
        if rain_forecast_mm > 50:
            print("⏳ ดินแห้ง แต่ฝนกำลังจะตกหนัก... ชะลอการรดน้ำ")
            pump.off()
        else:
            print("🌊 อนุมัติการรดน้ำ: เริ่มทำงานระบบไฮดรอลิก")
            pump.on()
    else:
        print("✅ ดินมีความชื้นพอเพียงแล้ว... ปิดระบบ")
        pump.off()

# ทดสอบลอจิกจำลอง
auto_irrigation(35, 10)
```
