Implementing threshold-based AI logic and rule-based decision making directly on the device for smart irrigation.

```python
def smart_irrigation_logic(moisture, is_raining, forecast_rain):
    # Simple Rule-based Edge AI
    if moisture < 30:
        if is_raining or forecast_rain > 70:
            return "WAIT" # Save water, rain is coming!
        else:
            return "WATER_ON"
    return "IDLE"

action = smart_irrigation_logic(25, False, 10)
print(f"Device Decision: {action}") # Should trigger watering
```
