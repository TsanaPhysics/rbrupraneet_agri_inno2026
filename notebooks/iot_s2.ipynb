หัวใจของ Smart Farm คือการเข้าใจว่าพืชต้องการอะไร ในเซสชันนี้เราจะเชื่อมต่อเซ็นเซอร์วัดความชื้นในดิน (Soil Moisture) และเซ็นเซอร์บรรยากาศ เพื่อแปลงกระแสไฟฟ้าให้เป็นข้อมูลที่โปรแกรมเข้าใจครับ

**เป้าหมาย:** 
1. เรียนรู้เรื่องสัญญาณ Analog vs Digital
2. ใช้งานโมดูล ADC (Analog to Digital Converter) บน ESP32
3. ฝึกการทำ Calibration เพื่อให้ได้ค่าเปอร์เซ็นต์ที่แม่นยำ

```python
from machine import ADC, Pin
import time

# เชื่อมต่อเซ็นเซอร์ความชื้นที่ขา 34 (Analog input)
soil_sensor = ADC(Pin(34))

# ตั้งค่าระดับการอ่านแรงดัน (0-3.3V)
soil_sensor.atten(ADC.ATTN_11DB)

def get_moisture_percent():
    # อ่านค่าดิบจากเซ็นเซอร์ (0-4095)
    raw_val = soil_sensor.read()
    
    # คำนวณเป็นเปอร์เซ็นต์ (สมมติ 4095 คือแห้งสนิท, 0 คือแช่น้ำ)
    percent = 100 - (raw_val / 4095.0 * 100)
    return percent

print("🔍 กำลังตรวจสอบระดับน้ำในดิน...")
moisture = get_moisture_percent()
print(f"💧 ค่าความชื้นปัจจุบัน: {moisture:.1f}%")
```
