Reading soil moisture sensors and mapping raw analog values to percentage levels for irrigation control.

```python
from machine import ADC, Pin

# 1. Setup Soil Moisture Sensor (Analog)
sensor = ADC(Pin(34))
sensor.atten(ADC.ATTN_11DB) # 0-3.6V range

def read_moisture():
    val = sensor.read()
    # Map 4095 (Dry) to 0% and 1500 (Wet) to 100%
    percent = ((4095 - val) / (4095 - 1500)) * 100
    return max(0, min(100, percent))

print(f"Current Soil Moisture: {read_moisture():.1f}%")
```
