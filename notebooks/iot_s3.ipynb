เซ็นเซอร์ที่ไม่ได้เชื่อมต่อกับอินเทอร์เน็ต ก็เหมือนกับโทรศัพท์ที่ไม่มีสัญญาณ ในตอนนี้เราจะมาเรียนรู้วิธีการทำให้ ESP32 คุยกับโลกภายนอกผ่านโปรโตคอล **MQTT** ซึ่งเป็นมาตรฐานระดับโลกสำหรับงานอุตสาหกรรมและเกษตรอัจฉริยะ

**เป้าหมาย:** 
1. เชื่อมต่อ ESP32 เข้ากับ Wi-Fi ของศูนย์นวัตกรรม
2. ทำความรู้จักกับ MQTT Broker (เซิร์ฟเวอร์กลางที่รับส่งข้อมูล)
3. ส่งข้อมูลความชื้นในดินขึ้นสู่ Dashboard แบบเรียลไทม์

```python
import network
import time
from umqtt.simple import MQTTClient

# 1. ตั้งค่าการเชื่อมต่อ Wi-Fi
SSID = 'RBRU-Inno-WiFi'
PASSWORD = 'AGRI-INNO-2026'

wlan = network.WLAN(network.STA_IF)
wlan.active(True)
wlan.connect(SSID, PASSWORD)

print("🌐 กำลังเชื่อมต่อ Wi-Fi...")
while not wlan.isconnected():
    time.sleep(1)
print("✅ เชื่อมต่ออินเทอร์เน็ตสำเร็จ! IP:", wlan.ifconfig()[0])
```

## 2. การส่งข้อมูลผ่าน MQTT
เราจะส่งข้อมูลไปที่ Topic `rbru/smart_orchard` เพื่อให้ระบบ Dashboard แสดงผลครับ

```python
SERVER = 'broker.hivemq.com'
CLIENT_ID = 'esp32_durian_001'
TOPIC = b'rbru/smart_orchard/moisture'

client = MQTTClient(CLIENT_ID, SERVER)

try:
    client.connect()
    print("📡 เชื่อมต่อ MQTT Broker สำเร็จ!")
    
    # จำลองการส่งข้อมูล
    data_to_send = "45.5"
    client.publish(TOPIC, data_to_send)
    print(f"📤 ส่งข้อมูล: {data_to_send} % ไปที่ระบบ")
    
except Exception as e:
    print("❌ ผิดพลาด:", e)
```
