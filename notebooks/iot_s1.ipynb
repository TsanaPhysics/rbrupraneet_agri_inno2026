Getting started with the ESP32 microcontroller and writing our first code in MicroPython.

```python
from machine import Pin
import time

# 1. Setup Built-in LED (usually GPIO 2)
led = Pin(2, Pin.OUT)

print("Starting blink test...")
for i in range(5):
    led.on()
    time.sleep(0.5)
    led.off()
    time.sleep(0.5)

print("Hardware Test Successful! 🎉")
```
