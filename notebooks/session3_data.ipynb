## 1. การเตรียมตัวเป็นนักสืบคอมพิวเตอร์
เราจะเริ่มจากการดาวน์โหลดไฟล์ชุดข้อมูลสภาพภูมิอากาศของจันทบุรี และใช้ Pandas เปิดอ่านครับ

```python
import pandas as pd

# ดาวน์โหลดไฟล์ตัวอย่างข้อมูลสภาพอากาศ (สามารถใช้ URL ของ GitHub ได้)
url = "https://raw.githubusercontent.com/TsanaPhysics/rbrupraneet_agri_inno2026/main/data/chanthaburi_climate_sample.csv"

try:
    df = pd.read_csv(url)
    print("✅ โหลดข้อมูลสำเร็จแล้ว!")
    # ดู 5 แถวแรกของข้อมูล
    display(df.head())
except:
    print("❌ ไม่พบไฟล์! (หากรันในเครื่องส่วนตัว อย่าลืมเช็ค Path ไฟล์นะครับ)")
```

## 2. การค้นหาความจริงในตัวเลข (Data Filtering)
เรามาลองหานะครับว่า ในรอบหลายปีที่ผ่านมา มีเดือนไหนบ้างที่เกิดปรากฏการณ์ **"Heavy Rain"** หรือฝนตกหนักผิดปกติ ซึ่งส่งผลต่อการออกดอกของทุเรียน

```python
# คัดกรองเฉพาะเดือนที่ Status เป็น Heavy Rain
heavy_rain_mask = df['Status'] == 'Heavy Rain'
df_heavy = df[heavy_rain_mask]

print("📅 เดือนที่มีปริมาณฝนวิกฤต (Heavy Rain Events):")
display(df_heavy[['Year', 'Month', 'Rainfall_mm']])

# คำนวณค่าฝนเฉลี่ยสูงสุด
max_rain = df['Rainfall_mm'].max()
print(f"\n🔥 บันทึกปริมาณฝนสูงสุดที่เคยพบ: {max_rain} มม.")
```

## 3. สรุปภาวะภูมิอากาศแยกตามปี
เรามาดูกันว่า ปีไหนที่จันทบุรีมีอากาศร้อนที่สุดเฉลี่ยทั้งปี

```python
yearly_summary = df.groupby('Year')['Avg_Temp'].mean()
print("🌡️ อุณหภูมิเฉลี่ยแยกตามปี:")
print(yearly_summary)
```
