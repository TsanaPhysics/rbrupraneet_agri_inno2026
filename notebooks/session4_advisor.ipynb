## 1. การสรุปข้อมูลด้วยภาพ (Charting Trends)
ก่อนจะรันระบบที่ปรึกษา เรามาวาดกราฟเส้นเพื่อดูเทรนอุณหภูมิของจันทบุรีในรอบหลายปีกันก่อนครับ

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/TsanaPhysics/rbrupraneet_agri_inno2026/main/data/chanthaburi_climate_sample.csv"
df = pd.read_csv(url)

# กรองข้อมูลเฉพาะเดือนเมษายน (เดือนที่ร้อนที่สุด) ของทุกปี
april_data = df[df['Month'] == 'Apr']

plt.figure(figsize=(10, 5))
plt.plot(april_data['Year'], april_data['Avg_Temp'], marker='o', linestyle='-', color='orange')
plt.title('April Temperature Trends in Chanthaburi (2016-2025)')
plt.xlabel('Year')
plt.ylabel('Temp (C)')
plt.grid(True)
plt.show()
```

## 2. พัฒนาระบบ Agri-Advisor ลอจิกอัจฉริยะ
เราจะสร้างโปรแกรมที่รับค่า "ปริมาณฝนที่คาดการณ์" และ "ระยะการเจริญเติบโตของทุเรียน" แล้วคืนค่าเป็นคำแนะนำครับ

```python
def smart_advisor(rain_mm, stage):
    """
    ฟังก์ชันช่วยตัดสินใจให้เกษตรกร
    Stage: 'flowering' (ออกดอก), 'fruitSet' (ติดผล), 'harvest' (เก็บเกี่ยว)
    """
    advice = "✅ สถานะ: ปกติ ดูแลตามร่องน้ำปกติ"
    
    if stage == 'flowering':
        if rain_mm > 50:
            advice = "⚠️ วิกฤต: ฝนชะดอก! รีบจัดทำทางระบายน้ำและเตรียมสารป้องกันเชื้อรา"
        elif rain_mm > 20:
            advice = "🔔 แจ้งเตือน: มีฝนปานกลาง ระวังความชื้นในช่อดอกสูงเกินไป"
            
    elif stage == 'harvest':
        if rain_mm > 0:
            advice = "🍊 ระวัง: ฝนตกช่วงเก็บเกี่ยวอาจทำให้คุณภาพเนื้อทุเรียนลดลง (ทุเรียนน้ำนิ่ง)"
            
    return advice

# --- ลองรันระบบจำลอง ---
current_prediction = 65.5 # ฝนที่จะตกพรุ่งนี้
durian_stage = 'flowering'

print("📡 --- RBRU Agri-Smart Advisor Report ---")
print(f"พยากรณ์ฝน: {current_prediction} mm")
print(f"ระยะพืช: {durian_stage}")
print(f"AI แนะนำ: {smart_advisor(current_prediction, durian_stage)}")
```
