**เป้าหมาย:** 
1. เข้าใจโครงสร้างของภาพดิจิทัล (Pixels & Channels)
2. ใช้งาน Library `OpenCV` เพื่อเปิดและปรับแต่งภาพใบพืช
3. เรียนรู้การทำ Grayscale และ Thresholding เพื่อเน้นเส้นใยใบไม้

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request

# ดาวน์โหลดภาพตัวอย่างใบพืช
url = 'https://raw.githubusercontent.com/TsanaPhysics/rbrupraneet_agri_inno2026/main/assets/images/research/leaf_sample.jpg'
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
img = cv2.imdecode(image, cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.imshow(img_rgb)
plt.title('Original Plant Leaf (RGB)')
plt.axis('off')
plt.show()
```

## 1. การลดความซับซ้อน (Grayscale Conversion)
ในการวิเคราะห์โรคพืช บางครั้งสีอาจสีเพี้ยนได้ตามแสง เราจึงต้องเน้นที่ "ความเข้มสว่าง" และ "รูปร่าง" ของแผลบนใบ

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Analysis')
plt.axis('off')
plt.show()
```
